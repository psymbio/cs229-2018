```python
# https://stats.stackexchange.com/questions/68391/hessian-of-logistic-function

# https://numpy.org/doc/stable/reference/routines.linalg.html

class LogisticRegression(LinearModel):
    """Logistic regression with Newton's Method as the solver.

    Example usage:
        > clf = LogisticRegression()
        > clf.fit(x_train, y_train)
        > clf.predict(x_eval)
    """

    def fit(self, x, y):
        """Newton's Method to minimize J(theta) for logistic regression.
        """

        def h(theta, x):
            """h_\theta(x) = 1 / (1 + exp(-\theta^T x)).
            """
            return 1 / (1 + np.exp(-np.dot(x, theta.transpose())))

        def gradient(theta, x, y):
            """J(\theta) = -(1/m) X^T (y - h(\theta x))
            """
            m, n = x.shape
            return -1 / m * np.dot(x.transpose(), (y - h(theta, x)))

        def hessian(theta, x):
            """Hessian of J(theta) = 1/m (x^T * h(\theta x) * (1-h(\theta x)) * x)
            x^T D x derivation: https://stats.stackexchange.com/questions/68391/hessian-of-logistic-function
            """
            m, n = x.shape
            h_theta_x = np.reshape(h(theta, x), (-1, 1))
            return 1 / m * np.dot(x.T, h_theta_x * (1 - h_theta_x) * x)

        def next_theta(theta, x, y):
            """theta updation
            """
            return theta - np.dot(np.linalg.inv(hessian(theta, x)), gradient(theta, x, y))

        m, n = x.shape

        # Initialize theta
        if self.theta is None:
            # only taken up to 2 in utils.py
            self.theta = np.zeros(3)

        # Update theta using Newton's Method
        old_theta = self.theta
        new_theta = next_theta(old_theta, x, y)
        while np.linalg.norm(new_theta - old_theta, 1) >= self.eps:
            old_theta = new_theta
            new_theta = next_theta(old_theta, x, y)

        self.theta = new_theta

    def predict(self, x):
        """Make a prediction given new inputs x.
        """
        return np.dot(x,self.theta) >= 0
```
